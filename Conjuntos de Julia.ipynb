## Conjuntos de Julia

Consideramos la función cuadrática $f:\mathbb{C}\to \mathbb{C}$, definida por  $$f_c(z)=z^2+c,$$
donde $c$ es un número complejo. 

Para cada $z\in \mathbb{C}$, nos interesa estudiar el comportamiento de la sucesión 
$$z, ~~ f_c(z), ~~ f_c(f_c(z)), ...$$

El conjunto de puntos para los que esta sucesión es acotada, se conoce como el conjunto de Julia y se denota por $J_c$.

http://math.bu.edu/people/bob/papers.html

#### Ejemplo 1

Graficar el conjunto de Julia de la función  

$$f_{0.7i}(z)=z^2+0.7i.$$

```python
import numpy as np
import matplotlib.pyplot as plt


%pylab inline

xvalues = linspace(-2, 2, 5)   
yvalues = linspace(-2, 2, 5)  

xlen = len(xvalues)
ylen = len(yvalues)

c = complex(0, .7) 

def julia(z, c, maxiter):         
    for iteration in range(maxiter):
        z = (z*z) + c
        if abs(z) > 2:
            break
            pass
        pass
    return iteration 

atlas = empty((xlen,ylen))

for ix in range(xlen):
    for iy in range(ylen):
        zx = xvalues[ix]
        zy = yvalues[iy]
        z  = complex(zx,zy)
        atlas[ix,iy] = julia(z,c,80)
        pass
    pass

figsize(18,18)
    
plt.imshow(atlas.T, interpolation = 'nearest', cmap='cubehelix') 

plt.savefig('MiJulia.pdf',bbox_inches='tight')

#print(atlas)
```
